```python
#Import dataframe
import pandas as pd
import numpy as np
sales = pd.read_csv('sales.csv')
prices = pd.read_csv('prices.csv')
print(sales.info())
```

```python
#convert ordered_at, updated_at to datetime object (using pd.to_datetime)
sales['ordered_at'] = pd.to_datetime(sales['ordered_at'])
prices['updated_at'] = pd.to_datetime(prices['updated_at'])
print(sales.info())
```

```python
#Sort column datetime ascending before merging
sales_sort = sales.sort_values('ordered_at')
prices_sort = prices.sort_values('updated_at')
```

```python
#Merge 'asof' 2 dataframe sales & prices on datetime column by product_id
merge_data = pd.merge_asof(sales_sort, prices_sort, by = 'product_id', left_on='ordered_at', right_on='updated_at', direction='nearest')
merge_data.head(10)
```

```python
#Fill NA
merge_data['listed_price'] = np.where(merge_data['ordered_at'] >= merge_data['updated_at'],
                                         merge_data['new_price'], merge_data['old_price'])
merge_data.head(10)
```

```python
#Add a new column named revenue
merge_data['total_revenue'] = merge_data['new_price']*merge_data['quantity_ordered']
merge_data.head(10)
```

```python
#Group by to calculate total revenue of each product at each price
result = merge_data.groupby(['product_id', 'listed_price'], as_index=False)['total_revenue'].sum()
result
```
